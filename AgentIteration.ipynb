````python
def exec_code_args(codes,*args):
    exec_code(codes)
    func_name= re.findall(r'def\s+(\w+)\s*\(',codes)[0]
    #get func_name
    exec_code(f'x_={func_name}({args[0][0]},{args[0][1]})')
    return x_
````

````python
exec_code_args("""
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1""",[[1,2,3],2])
````

````python
import json
````

````python
def extract_json(string):
    x= string.content[0].text
    x=re.findall(r'```json(.*)```',x,flags=re.DOTALL)[0]
    return json.loads(x)
````

````python
# a= Chat(model,sp='you are a python software engineer assistant')
codes=  re.findall(r'<code>(.*)</code>',a('tell if it is possible to get the target sum using elements of the array using python use code xml tags',prefill='<code>').content[0].text,flags=re.DOTALL)[0].strip()
#find code


prompt='''
look at the input given to you, and give me 5 test cases which seem reasonable enough to you, 
output format will be in json,think in about unique test cases, and novel ones and boundary cases, and write the final test cases
example
[{'input':{'arr':[1,2,3],'target':4},'expected_output':1},{'input':{'arr':[1,2,4],'target':40},'expected_output':-1},.....]
explanation: in example 4 is the sum of 1 and 3 so expected outptut is 1, wheras in the second one 43 sum is not possible with any combination of the numbers so expectedoutput is -1
'''
b=Chat(model,sp='you are a software tester assistant')

x=b(codes+prompt)#given by claude
#now proceeds to iterate 5 times for each test case and check if it works fine or not.
x= extract_json(x)
for i in range(len(x)):
    #print(codes,'----'*10,i,'----'*10)
    #instead has all the test casses it needs to implement 
    js= x[i]
    args= [js['input'][a] for a in js['input'].keys()]
    # print('here')
    output=exec_code_args(codes,args)
    #print(output)
    if output==js['expected_output']: print(f'Test case {i} passed,{args},{output}')
    else:
        while output!=js['expected_output']:
            print(f'Failed for {args},{output},{js["expected_output"]}')
            #loop till it is corrected
            codes=re.findall(r'<code>(.*)</code>',a(f"There was a problem with the inputs and target as {js['input']['arr']},{js['input']['target']}, the error was {output}",prefill='<code>').content[0].text,flags=re.DOTALL)[0].strip()
            output=exec_code_args(codes,args)
            print(codes)
        print('Corrected')
````
